# **Exploratory Data Analysis**

# Основная теория

**Exploratory Data Analysis (Разведочный Aнализ Данных)** - это подход к анализу данных с использованием визуальных и статистических методов.

EDA относится к критическому процессу выполнения первоначальных исследований данных с целью:

*   обнаружения закономерностей
*   выявления отклонений и аномалий
*   выбора наиболее важных параметров 
*   проверки гипотез предположений

Этот процесс закладывает основу для последующих этапов выбора признаков и проектирования, а также обеспечивает прочную основу для построения хороших моделей ML. Часто говорят, что если ML - это ракетный двигатель, то топливо - это (высококачественные) данные, которые мы можем передавать в алгоритмы ML.


---

Виды признаков:
*   Бинарные, которые принимают только два значения. 
*   Категориальные (номинальные). Они имеют конечное количество значений(дни недели, города и т.д.).
*   Упорядоченные. В некоторой степени похожи на категориальные признаки. Разница между ними в том, что в таких признаках существует четкое упорядочивание категорий (степень образования, время суток).
*   Числовые (количественные). Это значения в диапазоне от минус бесконечности до плюс бесконечности, которые нельзя отнести к предыдущим трем типам признаков.


## Статистический анализ данных


Целью такого анализа является определение качества признаков и их прогнозируемой способности. Данный вид анализа обеспечивает более полное понимание данных и должны быть первым шагом в изучении любого набора данных.


Исследование данных проводится с трех различных точек зрения:

*   описательной
*   корреляционной
*   контекстуальной








### Описательный анализ (одномерный)

Описательный анализ обеспечивает понимание характеристик каждого признака набора данных.

### Корреляционный анализ (двумерный анализ)

Корреляционный анализ исследует взаимосвязь между двумя признаками, коррелируют они или нет. Этот анализ может быть проведен с двух точек зрения для различных возможных комбинаций:

*  Качественный анализ.
*  Количественный анализ.

Мотивация для проведения корреляционного анализа заключается в том, чтобы помочь определить:


*   Какие признаки не являются прогностическими с точки зрения корреляции с целевым значением. Особое внимание обычно требуется непредсказуемым признакам, чтобы выявить более сильную связь с целевой переменной.
*   Какие признаки содержат избыточную информацию, которая может быть заменена производными признаками. Включение их всех может привести только к увеличению спроса на ресурсы и не принесет никакой выгоды в процессе ML.





#### **качественный анализ**



Качественный анализ - это в первую очередь исследовательский анализ, который дает представление о проблеме и помогает разработать идеи или гипотезы для потенциальных количественных исследований.

Для его проведения требуется наличие описательной статистики зависимых числовых или категориальных признаков по каждому уникальному значению независимого категориального признака. Эта перспектива помогает интуитивно понять взаимосвязь между двумя признаками. Визуализации часто используются вместе с качественным анализом как более наглядный способ представления результата.

Как видно из примера, качественный анализ дает некоторое представление о взаимосвязи между признаками. Чтобы подтвердить статистическую значимость разницы, можно применить количественный анализ.

#### **количественный анализ**

В случае корреляционного анализа количественный анализ проводится с помощью проверки статистической гипотезы:


1.   предложена гипотеза о статистической взаимосвязи между двумя признаками
2.   предложенная гипотеза сравнивается с альтернативной, идеализированной нулевой гипотезой (которая предполагает, что между двумя атрибутами нет никакой связи).

Сравнение считается статистически значимым, если взаимосвязь между атрибутами будет маловероятной реализацией нулевой гипотезы в соответствии с пороговой вероятностью: уровнем значимости (p-value).

Соответствующий количественный анализ, который может быть проведен по признакам пола и руки. Для этого воспользуемся критерием χ-квадрат. 


Гипотеза: 


*   H0: пол и рука независимы 
*   H1: пол и рука не независимы

1. Вычислим ожидаемое распределение, предполагая, что H0 (нулевая гипотеза) выполняется. Результат будет выглядеть так:

2. Вычислим $χ^2$:

 $χ^2=\sum_{i=1}^{row}\sum_{j=1}^{column}\frac{(O_{ij}-E_{ij})^2}{E_{ij}}=\frac{(43-45)^2}{45}+ \frac{(44-42)^2}{42}+\frac{(9-7)^2}{7}+\frac{(4-6)^2}{6}=1.42$

где 

$O_{ij}$ - реальное значение

$E_{ij}$ - ожидаемое значение


3. $p-value = Pr_{DF=1}(\chi^2>1.42) >= 0.2$


[Values of the Chi-squared distribution](https://www.medcalc.org/manual/chi-square-table.php)





p-value меньше порогового значения (обычно 0,05 или 0,01), следовательно нулевая гипотеза отклоняется. Это фактически дает результат, противоречащий наблюдениям, сделанным в предыдущем разделе. Данный пример демонстрирует важность количественного анализа.

### Контекстуальный анализ

Поскольку ни описательный анализ, ни корреляционный анализ не требуют контекстной информации, оба они являются достаточно общими для выполнения с любым (структурированным) набором данных. Для дальнейшего понимания или профилирования данного набора данных и получения более подробной информации о предметной области рекомендуется использовать два общих анализа, основанных на контекстной информации:


*   **анализ основанный на времени (time-based analysis)**
*   **анализ основанный на агентах (agent-based analysis)**


Ожидается, что качество набора данных может быть дополнительно проверено на основе знаний предметной области из результатов контекстуального анализа.

####  Time-based analysis

Во многих реальных наборах данных временная метка (или аналогичные признаки, связанные со временем) является одним из ключевых элементов контекстуальной информации. Например, журналы операций для онлайн-службы API обычно содержат время создания журнала и/или время, когда произошло зарегистрированное событие. Журналы транзакций для розничной компании обычно содержат время, в которое произошла транзакция. Наблюдение и/или понимание характеристик данных во временном измерении с различной степенью детализации имеет важное значение для понимания процесса генерации данных и обеспечения качества данных.
С помощью атрибута timestamp можно выполнить следующие анализы:


*   Количество записей (транзакций) за интервал времени 

  *   Например, ежечасные просмотры страниц веб-сайта
*    Количество уникальных значений (признаков) за интервал времени
  *   Например, ежечасные уникальные посетители веб-сайта
*   Описательная статистика за интервал времени
  *   Например, среднесуточная продолжительность сеанса для веб-сайта


Временной интервал может быть в минутах, часах или днях, в зависимости от задачи.

#### Agent-based analysis

В качестве альтернативы метке времени другим распространенным атрибутом является уникальный идентификатор (ID) для каждой записи. Эти представляющие идентификаторы предоставляют важную контекстуальную информацию, включая, например:
*   Идентификатор пользователя в журнале транзакций активности пользователя
*   Идентификатор магазина и/или идентификатор товара в журнале транзакций для розничного магазина
*   Идентификатор события в журнале операций для потоковой системы


С помощью атрибута ID можно выполнить следующие анализы:


*   Гистограмма и/или описательная статистика количества записей на агента 
  *   Например, средняя транзакция на конкретного пользователя за X месяцев
*   Гистограмма и/или описательная статистика количества уникальных значений (атрибута) для каждого агента
  *   Например, среднее количество уникальных клиентов в конкретном магазине за X месяцев

## Выбор и проектирование признаков


Конечная цель EDA состоит в том, чтобы дать представление об изучаемом наборе данных. 

Очевидно, что проблемы в наборе данных могут быть выявлены только в результате строгого статистического анализа данных и/или интуитивно идентифицирован на основе визуализации с оценкой знаний предметной области.

### Выбор признаков с помощью описательного анализа

Описательный анализ предоставляет базовую статистику каждого атрибута набора данных, и на основе этого можно определить некоторые плохие или несущественные признаки.

Дальше рассмотрим идикаторы плохих признаков, на основе которых могут быть приняты соответствующие решения о том, что с такими признаками делать.











*   **Высокий процент пропущенных значений.** Выявленная проблема заключается в том, что значения признаков отсутствует в значительной части примеров данных. 

Есть два варианта решения этой проблемы, в зависимости от бизнес-сценария:
  1.  **Отсутствующее значение** в определенных контекстах **имеет смысл**. Например, оно может указывать на то, что отслеживаемый базовый процесс не функционирует должным образом. Поэтому **присвоение уникального значения** отсутствующим записям значений является хорошим решение такой проблемы.
  2.   Значение может **отсутствовать из-за неправильной настройки, проблем со сбором данных или не отслеживаемых случайных причин**, тогда **признак может быть выброшен** (если невозможно восстановить истинные пропущенные значения данного признака).

Важно понимать, что отстутсвие значений в записях можно разделить на 3 случая:

  *  связано с особенностями некоторых наблюдаемых данных, а не их "неправильным/плохим" сбором
  *  не имеет ничего общего с гипотетическим значением или со значениями других переменных.
  *   Недостающее значение зависит от гипотетического значения (например, люди с высокой зарплатой обычно не хотят раскрывать свои доходы в опросах) или недостающее значение зависит от значения какой-либо другой переменной (например, женщины обычно не хотят раскрывать свой возраст).

В первых двух случаях безопасно удалять данные с пропущенными значениями, в то время как в третьем случае удаление наблюдений с пропущенными значениями может привести к смещению модели.


*   **Низкая дисперсия числовых признаков.**Выявленная проблема заключается в очень малой дисперсии признака по сравнению с его типичным диапазоном значений, или распределение представляет собой резкую колоколообразную кривую. В большинстве случаев безопасно удалять числовые значения с низкой дисперсией. Это не повредит производительности модели и может снизить сложность модели.
*   **Низкая энтропия категориальных признаков.** Выявленная проблема заключается в очень малой энтропии объекта, что означает, что большинство записей имеют одинаковые категориальные значения. В большинстве случаев безопасно удалять категориальные признаки с низкой энтропией. Это не повредит производительности модели и может снизить сложность модели.


*   **Дисбаланс категориальной целивой переменной (классовый дисбаланс)**. Набор данных считается “сильно несбалансированным по классам”, если выборка из одного целевого класса значительно превышает по численности другие. Это можно рассматривать как частный случай предыдущей “низкой энтропии категориальных атрибутов”. В несбалансированном наборе данных класс с большим числом экземпляров называется основным классом, в то время как классы с относительно меньшим количеством экземпляров называются второстепенными классами.

В этом случае большинство классификаторов смещены в сторону основных классов и, следовательно, показывают низкие показатели классификации по второстепенным классам. Также возможно, что классификатор предсказывает все как основной класс и игнорирует второстепенный класс. Для устранения классового дисбаланса было предложено множество методов:

  1.  **Undersampling majority class.** Важный метод, который пытается сбалансировать распределение классов путем случайного удаления образцов из класса большинства (majority class). Основной проблемой, связанной с этим методом, является потеря ценной информации
  2.   **Oversampling minority class.** Методы случайной переборной выборки также помогают
достичь баланса в распределении классов путем воспроизведения выборки класса меньшинств (minority class). Однако при этом в набор данных не добавляется новая информация. Один из таких методов: ([SMOTE](https://jair.org/index.php/jair/article/view/10302))
  3.   **Альтернативная метрика и/или функция потерь.** Метрики AUC и F1 score (average of precision and recall) хорошо подходят для оценки полученных результатов для несбалансированных классов.Также возможно изменить функцию потерь, наложив дополнительные штрафы на модель за допущение ошибок классификации в классе меньшинства во время обучения. Эти штрафы могут заставить модель уделять больше внимания классу меньшинств.



*   **Асимметричное распределение.** Выявленная проблема заключается в том, что распределение числового признака имеет форму длинного хвоста.

Существует несколько вариантов решения этой проблемы, в зависимости от бизнес-сценария:


1.   Иногда выбросы могут быть вызваны неверными данными, и в этом случае понимание источника ошибки может позволить изменить значение выбросов на правдоподобное.
2.   Если экстремальные значения не могут рассматриваться как выбросы, обычно рекомендуются преобразования (например, логарифмические преобразования) для получения более сбалансированного распределения.
3.   Метод equal-frequency binning. Сгруппирование числового признака таким образом, чтобы количество точек данных в каждом сегменте или ячейке было сбалансировано. Этот метод может эффективно уменьшить перекос и сохранить способность прогнозирования.  Существуют и другие binning стратегии (например, binning равной ширины), которые не могут устранить перекос.

4.  экстремальные значения на самом деле вызваны выбросами. Отфильтровывание экстремальных значений или выбросов приведет к нормальному распределению признака. 

Существует два типичных метода удаления выбросов:
  *   Отсечение признака в вычисленном процентиле (например, 99%), предполагая, что большая часть данных верна и только небольшой процент (например, 1%) является ненормальным.
  *   Фильтрация по фиксированному порогу, если известно, что существует нормальный диапазон значений для числового признака (например, широта и долгота).



*   **High cardinality.** Когда количество уникальных значений слишком велико для категориальных признаков появляется проблема с one-hot-encoding процессом.
Типичным средством решения такой проблемы является:
1.   Применение [hash trick](https://en.wikipedia.org/wiki/Feature_hashing), который преобразует категориальный признак с большим кол-ом значений в пространство фиксированного размера.
2.   [Embedding](https://www.tensorflow.org/guide/embedding) - отображение дискретных объектов, таких как слова, в векторы действительных чисел.
3.   Удаление признаков, если количество вхождений на одно уникальное значение признака слишком мало (например, менее 5).


### Выбор признаков на основе корреляционного анализа

Виды проблем, которые можно выявить при таком виде анализа:

*   **Низкая корреляция между признаком и целевой переменной**. Возможны две причины:


1.  Этот признак не является полезным с точки зрения прогнозирования верного ответа, поэтому он может быть удален из исследования.
2.   Этот признак бесполезен, в текущем виде и требует преобразование, чтобы выявить более сильную связь с целью. Одним из возможных примеров является ситуация, когда долгота и широта не предоставляют четкой информации в необработанном виде. Их объединение для создания понятия “местоположение” часто является полезным преобразованием, которое может увеличить их корреляцию с целевой переменной.

По этим причинам, прежде чем удалять функции из области видимости, рекомендуется, чтобы эксперт по предметной области рассмотрел результаты анализа, чтобы принять окончательное решение.

*  **Высокая корреляция между признаками.** Наличие сильно коррелированных признаков может быть проблемой, потому что:


1.   Никакой дополнительной информации не предоставляется, и она не помогает улучшить производительность модели.
2.   Требования к вычислительным мощностям увеличиваются во время обучения.
3.   Некоторые модели (например, линейные) могут быть нестабильными.

Типичные методы решения этой проблемы включают:


1.   Удаление всех, кроме одного из сильно коррелированных признаков.
2.   Применение уменьшения размерности.
3.   Использование нелинейной модели (например, нейронной сети), которая обычно является более надежной при наличии коррелированных признаков.







### Выбор функций на основе контекстуального анализа

Основная цель предоставления контекстуального анализа - помочь пользователям лучше понять данные. Визуализация данных во временных и/или агентских измерениях должна помочь в понимании их контекста. Однако действия, которые могут быть предприняты в этом случае, в большей степени зависят от знаний разработчика о предметной области.

Таким образом, достаточно просто представить визуализацию и оставить окончательное решение за разработчиком.
